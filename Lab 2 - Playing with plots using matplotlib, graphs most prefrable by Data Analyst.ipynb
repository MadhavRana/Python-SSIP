# Program 1 - Plot Colors, Markers and line styles in MatplotLib

# Plot Marker

 

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o')
plt.show()
```

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o', ms = 20, mec = 'b')
plt.show()
```

# Line Style

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, linestyle = 'solid')
plt.show()
```

# Color

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, color = 'black')
plt.show()
```

# Program 2 - Different Plots in MatplotLib

# Line Plot

```python
from matplotlib import pyplot as plt

# x-axis values
x = [5, 2, 9, 4, 7]

# Y-axis values
y = [10, 5, 8, 4, 2]

# Function to plot
plt.plot(x, y)

# function to show the plot
plt.show()
```

```python
x = np.arange(0, 3 * np.pi, 0.1) 
print(x)
y = np.sin(x) 
plt.title("sine wave form") 

# Plot the points using matplotlib 
plt.plot(x, y) 
plt.show()
```

# Bar Chart

```python
from matplotlib import pyplot as plt

# x-axis values
x = [5, 2, 9, 4, 7]

# Y-axis values
y = [10, 5, 8, 4, 2]

# Function to plot
plt.bar(x, y)

# function to show the plot
plt.show()
```

```python
from matplotlib import pyplot as plt

# x-axis values
x = [5, 2, 9, 4, 7, 3, 8, 10, 1]

# Y-axis values
y = [10, 5, 8, 4, 2, 1, 6, 3, 9]

# Function to plot scatter
plt.scatter(x, y)

# function to show the plot
plt.show()
```

# Pie Chart

```python
cars = ['AUDI', 'BMW', 'FORD',
        'TESLA', 'JAGUAR', 'MERCEDES']

data = [23, 17, 35, 29, 12, 41]
explode = [0,0.3,0,0,0,0]
# Creating plot
fig = plt.figure(figsize =(10, 7))
plt.pie(data, labels = cars,explode=explode)


# show plot
plt.show()
```

# Histogram and Binning



```python
data = np.random.normal(170, 10, 250)
#The hist() function will read the array and produce a histogram:
#set the bin value 30
plt.hist(data);

plt.show() 
plt.hist(data, bins=30);

plt.show() 
```

# Text and Annotation

```python
fig, ax = plt.subplots()

x = np.linspace(0, 20, 1000)
ax.plot(x, np.cos(x))
ax.axis('equal')
#text annotation to the plot where it indicate maximum value of the curv
ax.annotate('local maximum', xy=(6.28, 1), xytext=(10, 4),
            arrowprops=dict(facecolor='black', shrink=0.05))
#text annotation to the plot where it indicate minimum value of the curv
ax.annotate('local minimum', xy=(5 * np.pi, -1), xytext=(2, -6),
            arrowprops=dict(arrowstyle="->",
                            connectionstyle="angle3,angleA=0,angleB=-90"));
```

# Legends
